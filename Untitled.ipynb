```python
import pandas as pd
```

## Criando DataFrame com um dicionário

```python
venda = {'data':['18/08/2022', '19/08/2022'],
         'valor':[500, 300],
         'produto':['feijão', 'arroz'],
         'quantidade': [50,70],
        }
vendas_df = pd.DataFrame(venda)
```

```python
display(vendas_df)
```

## Criando DataFrame importando dados do Excel

```python
vendas_df = pd.read_excel("vendas.xlsx")
display(vendas_df)
```

## Métodos de Visualização 

#### Método Head() - Mostra as 5 primeiras linhas ou dentro do parênteses pode especificar quantas linhas visualizar 

```python
display(vendas_df.head())
```

#### Método Shape - Mostra apenas quantidade de linhas e colunas

```python
display(vendas_df.shape)
```

#### Método Describe() - Mostra Resumo (Quantidade de linhas, Média, Desvio padrão, Mínimo, Máximo, Quartis )

```python
display(vendas_df.describe())
```
